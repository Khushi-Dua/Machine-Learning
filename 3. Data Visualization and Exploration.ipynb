```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Load the Titanic Dataset
The Titanic dataset in Seaborn is a built-in dataset used for classification tasks, specifically to predict passenger survival on the Titanic.
### Key features:
- **Survived**: Whether the passenger survived (0 = No, 1 = Yes).
- **Pclass**: Passenger class (1, 2, 3).
- **Sex**: Gender (male, female).
- **Age**: Age of the passenger.
- **SibSp**: Number of siblings/spouses aboard.
- **Parch**: Number of parents/children aboard.
- **Fare**: Fare paid by the passenger.
- **Embarked**: Embarkation port (C = Cherbourg, Q = Queenstown, S = Southampton).

```python
# The Titanic dataset can be directly loaded from seaborn's built-in datasets
df = sns.load_dataset('titanic')
```

```python
# Display the first few rows of the dataset
print("First few rows of the Titanic dataset:")
print(df.head())
```

```python
# Display basic statistics of the dataset
print("\nBasic statistics of the Titanic dataset:")
print(df.describe(include='all'))
```

# 2. Create histograms, scatter plots, box plots, and heatmaps to visualize different features.

## Histograms

```python
# Create histograms for numerical features
plt.figure(figsize=(14, 6))
df.hist(bins=20, figsize=(14, 6), layout=(2, 4))
plt.tight_layout()
plt.show()
```

## Scatterplots

```python
# Create scatter plots for numerical features
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='Set1')
plt.show()
```

## Boxplots

```python
# Create box plots for categorical features vs numerical features
plt.figure(figsize=(12, 6))
sns.boxplot(x='pclass', y='age', data=df)
plt.title('Box plot of Age vs Pclass')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Box plot of Fare vs Pclass')
plt.show()
```

## Heatmap

```python
# Select only numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Generate the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlations between Numerical Features')
plt.show()
```
