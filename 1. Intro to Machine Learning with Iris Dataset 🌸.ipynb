# AIM: Write a Python program to load a dataset (e.g., Iris dataset), display basic statistics, and plot a simple graph.

The Iris dataset is a classic dataset used for machine learning, available at Scikit-learn (load_iris()), containing 150 samples of iris flowers from three species: **Setosa**, **Versicolor**, and **Virginica**. Each sample has 4 features: `sepal length`, `sepal width`, `petal length`, and `petal width` **(all in cm)**. The dataset is commonly used for classification tasks to predict the species of a flower based on its features.


### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Iris Dataset

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

### Convert to DataFrame

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

### Add the target (species) as a new column
In the Iris dataset:

- 0 is assigned to Setosa,
- 1 is assigned to Versicolor,
- 2 is assigned to Virginica.
This encoding is done automatically when using the Scikit-learn's `load_iris()` function,which stores the target species as integers:

- `iris.target`: Contains the integer-encoded species labels, where each integer corresponds to a different flower species.

```python
df['species'] = iris.target
```

```python
print("BASIC STATISTICS OF IRIS DATASET:")
print(df.describe())
```

## PLOTTING SIMPLE GRAPH

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'], c=df['species'], cmap='viridis', edgecolors='k', s=100)
plt.title("Sepal length VS Sepal Width")
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Sepal Width (cm)')
#Show color legend
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=10, label='Setosa'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Versicolor'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Virginica')],
           title="Species", loc="best")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['petal length (cm)'],df['petal width (cm)'], c=df['species'], cmap='viridis', edgecolors='k', s=100)
plt.title("Petal length VS Petal Width")
plt.xlabel('Petal Length(cm)')
plt.ylabel('Petal Width (cm)')
#Show color legend
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=10, label='Setosa'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Versicolor'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Virginica')],
           title="Species", loc="best")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['petal length (cm)'],df['sepal length (cm)'], c=df['species'], cmap='viridis', edgecolors='k', s=100)
plt.title("Petal length VS Sepal length")
plt.xlabel('Petal length (cm)')
plt.ylabel('Sepal Length(cm)')
#Show color legend
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=10, label='Setosa'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Versicolor'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Virginica')],
           title="Species", loc="best")
#plt.Line2D([0], [0], ...) is used to create a line object in Matplotlib, but in this case, it's used to create markers (such as circles) for the legend, not actual lines.

# Show the plot
plt.show()
```
